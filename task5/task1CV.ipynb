```python
point1 = None
point2 = None
box = None

def cropFunction(event,x,y,flags,param):
    global point1, point2, box
    if event == cv2.EVENT_LBUTTONDOWN:
        if point1 is None:
            point1 = (x,y)
            print(f"Point 1: {point1}")
            
        elif point2 is None:
            point2 = (x,y)
            print(f"Point 2: {point2}")
             
            
            minY= min(point1[1],point2[1])
            maxY= max(point1[1],point2[1])

            minX=min(point1[0],point2[0])
            maxX=max(point1[0],point2[0])
            box=cv2.rectangle( cv2.imread("F:/computer vision/skyline.jpg", 1), (maxX,maxY), (minX,minY), (0,0,255))
         
# Creating a window to add the mouse callback to

cv2.namedWindow('image')
cv2.setMouseCallback('image',cropFunction)
image=cv2.imread("F:/computer vision/skyline.jpg", 1)

while(1):
    cv2.imshow('image',image)
    if box is not None:
        cv2.imshow('image',box)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
        
```
